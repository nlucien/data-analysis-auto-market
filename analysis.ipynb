### 📂 Dataset

```python
import pandas as pd

data = {
    "marque": ["Toyota","Nissan","Mercedes","Kia","Hyundai","BMW","Toyota","BMW","Audi","Peugeot"],
    "prix": [8500000,7200000,15000000,4300000,9900000,12000000,7800000,20000000,11000000,6500000],
    "annee": [2018,2017,2022,2015,2020,2021,2019,2023,2020,2016],
    "kilometrage": [75000,92000,15000,120000,40000,22000,60000,5000,30000,110000],
    "carburant": ["Essence","Diesel","Essence","Diesel","Essence","Diesel","Essence","Essence","Diesel","Essence"]
}

df = pd.DataFrame(data)
```

### 🧠 PARTIE 3 — Analyse Avancée (6 pts)

#### ✅ Créer une colonne :

#### prix_par_km = prix / kilometrage

#### Puis :

#### ✅ Trouver la meilleure affaire (prix_par_km minimum)
#### ✅ Trier les voitures du plus récent au plus ancien

```python
# création du nouvelle colonne prix_par_km
df["prix_par_km"] = df["prix"] / df["kilometrage"]
# Trouvons la meilleure affaire prix_par_km_min
prix_par_km_min = df.loc[df["prix_par_km"].idxmin()]
# Trions les voitures du plus récent au plus ancien
voitures_plus_recent_au_plus_ancien = df.sort_values("prix_par_km", ascending=False)
```

### 🧠 PARTIE 4 — Analyse Business (7 pts)

#### Fais :

#### ✅ Prix moyen par marque
#### ✅ Prix moyen par type de carburant

#### Puis :

#### 👉 Réponds en texte :

#### 1️⃣ Quelle variable influence le plus le prix ?
#### 2️⃣ Y a-t-il une relation entre année et kilométrage ?
#### 3️⃣ Est-ce un marché premium ou économique ?

```python
# Prix moyen par marque
prix_moyen_par_marque = df.groupby("marque")["prix"].mean()
# Prix moyen par type de carburant
prix_moyen_par_type_de_carburant = df.groupby("carburant")["prix"].mean()

# Réponse en texte
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix)
# 1- La variable qui influence plus le prix est l'année car la valeur absolue de la relation annee et prix est supérieur à la valeur absolue de la relation prix et les autres variables.
# 2- oui il y a une forte relation entre année et kilométrage qui est de - 0.98.
# 3- Un marché premium ou économique, pour le savoir il me faut faire un histogramme.
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df["prix"], bins=5)
plt.title("Distribution des prix de voitures")
plt.xlabel("Prix")
plt.ylabel("Nombre de voiture")
plt.show()
# Après la visualisation de l'histogramme, je conclus que c'est un marché économique
```
